```python
genres = []
for genre in dataset.genres:
    x = genre.split('|')
    for i in x:
        if i not in genres:
            genres.append(str(i))
genres = str(genres)
movie_title = []
for title in dataset.title:
    movie_title.append(title[0:-7])
movie_title = str(movie_title)    
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
wordcloud_genre = WordCloud(width=1400, height=900, background_color='black', min_font_size=2,
                           min_word_length=3).generate(genres)
wordcloud_title = WordCloud(width=1400, height=900, background_color='black', min_font_size=2,
                           min_word_length=3).generate(movie_title)
```

```python
plt.figure(figsize=(30,15))
plt.axis('off')
plt.title('Movie Genres Wordcloud', fontsize=25)
plt.imshow(wordcloud_genre)
```

```python
plt.figure(figsize=(30,15))
plt.axis('off')
plt.title('Movie Title Wordcloud', fontsize=25)
plt.imshow(wordcloud_title)
```

```python
dataset1 = pd.merge(rating, dataset, how='left', on='movieId')
dataset1.head()
```

```python
dataset2 = dataset1.groupby(['title'])[['rating']].sum()
high_rated = dataset2.nlargest(20, 'rating')
high_rated.head()
```

```python
plt.figure(figsize=(30,10))
plt.title('Top 20 movies with highest rating',fontsize=50)
colors=['red','yellow','orange','green','magenta','cyan','blue','lightgreen','skyblue','purple']
plt.ylabel('ratings',fontsize=30)
plt.xticks(fontsize=25,rotation=90)
plt.xlabel('movies title',fontsize=30)
plt.yticks(fontsize=25)
plt.bar(high_rated.index,high_rated['rating'],linewidth=3,edgecolor='red',color=colors)
```

```python
dataset3 = dataset1.groupby(['title'])[['rating']].count()
rating_count = dataset3.nlargest(20, 'rating')
rating_count.head()
```

```python
plt.figure(figsize=(30,10))
plt.title('Top 20 movies with highest number of ratings',fontsize=30)
plt.xticks(fontsize=25,rotation=90)
plt.yticks(fontsize=25)
plt.xlabel('movies title',fontsize=30)
plt.ylabel('ratings',fontsize=30)

plt.bar(rating_count.index,rating_count.rating,color='red')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
indices=pd.Series(dataset.index,index=dataset['title'])
titles=dataset['title']
def recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:21]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```
